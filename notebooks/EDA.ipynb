INSIGHTS

Category-wise Performance


Technology category contributes the highest sales.

Office Supplies category contributes the lowest sales.

Regional Profit Comparison


The West region generates the highest profit.

The Central region shows the lowest profit.

Segment-wise Profitability


The Consumer segment generates more profit.

The Home Office segment reports lower profit.

Sales Trend Insights


Sales are increasing year by year, as observed from the monthly sales trend.

Correlation-Based Insights


Sales and profit are positively correlated — when sales increase, profit tends to increase.

Discount and profit are negatively correlated — higher discounts generally reduce profit.

Quantity sold and sales have a slight positive correlation — selling more units slightly increases sales.
